```python
def fasta_csv(fasta_dir,temp_file_path,output_dir):
    file_name_list = os.listdir(fasta_dir)
    protein_list=[]
    for file in file_name_list:
        abs_file_path = fasta_dir+file
        # print(abs_file_path)
        f = open(abs_file_path,'r')
        lines = f.read().splitlines()
        protein_list.append(lines[1])
        f.close()
    p_dict ={"Info_window_seq":protein_list} 
    df = pd.DataFrame(p_dict)
    
    df.to_csv(temp_file_path,index=False)
    # df = pd.read_csv()
    pep.aa_all_feat.calc_csv(dataframe=df, save_folder=output_dir,aa_column='Info_window_seq'
                                     ,Ncores=1,chunksize=None, k=2)

        
```

```python
fasta_csv(fasta_dir,temp_file,output_dir)
```
